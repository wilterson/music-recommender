#### Será feita uma análise do gênero da última música curtida pelo usuário e com base nisso, pega-se as características desse gênero e realiza-se uma similaridade de cosseno para encontrar as 30 músicas mais parecidas. Dessas 30, ordena-se as 10 com maior popularidade. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pltd
import seaborn as sns
import pickle
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#importando arquivos
df_data = pd.read_csv('https://drive.google.com/uc?export=download&id=1tdnYgbK3-OJeeMK-QrW-ha6WgXcC79hE', low_memory=False, encoding='latin-1')
df_data_by_genres = pd.read_csv('https://drive.google.com/uc?export=download&id=12lqn5lD6x5Yh6mezlg6TQ60YA-Pn8aeV', low_memory=False, encoding='latin-1')
```

```python
#removendo duplicatas do arquivo que contém as músicas
name_artist = ['name','artists']
df_data = df_data.drop_duplicates(subset=name_artist,keep='last')
```

```python
#colunas de interesse do arquivo que contém as músicas
df_music = df_data[['acousticness','danceability','energy','instrumentalness', 'liveness', 'loudness', 'speechiness','valence']]
```

```python
#analisando o arquivo com as características dos gêneros
df_data_by_genres.head(3)
```

```python
#função que retorna as músicas similares e com maior popularidade
def get_similarities_by_genre(genero):
    rock_songs = df_data_by_genres[df_data_by_genres['genres'] == genero]
    input_vector = rock_songs[['acousticness','danceability','energy','instrumentalness', 'liveness', 'loudness', 'speechiness','valence']]
    cosine_similarities = cosine_similarity(df_music, input_vector)
    df_retorno_genres = df_music.iloc[cosine_similarities[:,0].argsort()[::-1][:30]].merge(df_data[['id','name','artists','popularity']],how = 'inner',left_index=True,right_index=True).sort_values('popularity',ascending=False).head(10)
    return df_retorno_genres
```

```python
#biblioteca picke para criar um pipeline
pickle.dump(get_similarities_by_genre,open('similarity.pkl','wb'))
```
